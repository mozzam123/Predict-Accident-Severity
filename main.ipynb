#### Drop Unwanted Columns

```python
columns_to_drop = ['ID','Description','Street','City','County','State','Zipcode','Country','Source','Amenity','Bump','Crossing', 'Give_Way', 'Junction', 'No_Exit', 'Railway',
    'Roundabout', 'Station', 'Stop', 'Traffic_Calming', 'Traffic_Signal','Turning_Loop', 'Timezone' , 'Airport_Code', 'Sunrise_Sunset', 'Civil_Twilight', 'Nautical_Twilight', 'Astronomical_Twilight','Weather_Timestamp']

df = df.drop(columns=columns_to_drop)
```

```python
df.columns
```

```python
df.tail()
```

```python
# Removind 'End_lat', 'Eng_lag' and 'Precipitation(in)' as there are too many null values, may not add value
df = df.drop(columns=['End_Lat', 'End_Lng', 'Precipitation(in)', 'Wind_Chill(F)'])
df.columns
```

```python
# convert start_time to float64
df['Start_Time'] = pd.to_datetime(df['Start_Time'], format='mixed')
df['End_Time'] = pd.to_datetime(df['End_Time'], format='mixed')
```
